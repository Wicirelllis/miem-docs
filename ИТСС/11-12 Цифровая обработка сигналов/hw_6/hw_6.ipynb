## Предисловие

Группы ячеек для каждого задания раскрываются.

Графики - диаграммы Боде [wiki](https://en.wikipedia.org/wiki/Bode_plot) (ЛАФЧХ, Логарифмическая амплитудно-фазовая частотная характеристика(!)).
Это парные графики, верхний - АЧХ, нижний - ФЧХ.
Горизонтальная ось - частота, причем в лог-масштабе.
Вертикальная - усилене в дБ для АЧХ, сдвиг фазы в градусах для ФЧХ.

Параметры типо как в ```signal.cheby1(4, 5, 100, btype="lowpass", analog=False, output="ba", fs=10_000), dt=0.05)))``` брал просто из примеров документации сайпая.

````python
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt
import numpy as np
from scipy import fft, signal
````

````python
def make_plot(title: str, w: list, mag: list, phase: list) -> None:
    subtitle = ["lowpass", "highpass", "bandpass", "bandstop"]
    fig = plt.figure(figsize=(10, 8), dpi=130)
    outer = gridspec.GridSpec(2, 2)

    for i in range(4):
        inner = gridspec.GridSpecFromSubplotSpec(2, 1, subplot_spec=outer[i])

        ax = plt.Subplot(fig, inner[0])
        ax.set_title(subtitle[i])
        ax.semilogx(w[i], mag[i])
        ax.grid(which='both', axis='both')
        fig.add_subplot(ax)

        ax = plt.Subplot(fig, inner[1])
        ax.semilogx(w[i], phase[i])
        ax.grid(which='both', axis='both')
        fig.add_subplot(ax)

    fig.suptitle(title)
    fig.tight_layout()
````

## Задание 3
Построить нули и полюса цифровых фильтров.

Кружочки - нули, крестики - полюса.
Также я добавил серую единичную окружность для более понятного масштаба.

````python
def zero_pole_plot(title: str, z: list, p: list) -> None:
    subtitle = ["lowpass", "highpass", "bandpass", "bandstop"]
    fig = plt.figure(figsize=(10, 8), dpi=130)
    outer = gridspec.GridSpec(2, 2)

    for i in range(4):
        ax = plt.Subplot(fig, outer[i])
        ax.axis('equal')
        ax.set_title(subtitle[i])
        cir = plt.Circle((0, 0), 1, color="grey", fill=False)
        ax.add_patch(cir)
        ax.plot(z[i].real, z[i].imag, "o")
        ax.plot(p[i].real, p[i].imag, "x")
        ax.grid(which='both', axis='both')
        fig.add_subplot(ax)

    fig.suptitle(title)
    fig.tight_layout()
````

````python
data = []
data.append(signal.butter(4, 100, btype="lowpass", analog=False, output="zpk", fs=10_000))
data.append(signal.butter(4, 200, btype="highpass", analog=False, output="zpk", fs=10_000))
data.append(signal.butter(4, [100, 1000], btype="bandpass", analog=False, output="zpk", fs=10_000))
data.append(signal.butter(4, [100, 1000], btype="bandstop", analog=False, output="zpk", fs=10_000))

z = [i[0] for i in data]
p = [i[1] for i in data]
zero_pole_plot("Butterworth, digital", z, p)
````

## Задание 5
Применить фильтры, рассмотренные выше, к сигналу - последовательности прямоугольных импульсов.

````python


t = np.linspace(0, 1./100, 300)
sig = signal.square(2 * np.pi * 300 * t)
fig, ax = plt.subplots(2, dpi=150)
fig.autofmt_xdate(rotation=45)
ax[0].plot(t, sig)
ax[0].set_title("Square wave")

x_ = fft.fftfreq(len(sig), 1/300)
y_ = np.abs(fft.fft(sig)) / len(t)
x, y = zip(*sorted(zip(x_, y_)))
ax[1].plot(x, y)
ax[1].set_title("Spectrum")
````

````python
ba = []
ba.append(signal.butter(4, 1000, btype="lowpass", analog=False, output="ba", fs=10_000))
ba.append(signal.butter(4, 2000, btype="highpass", analog=False, output="ba", fs=10_000))
ba.append(signal.butter(4, [100, 1000], btype="bandpass", analog=False, output="ba", fs=10_000))
ba.append(signal.butter(4, [100, 1000], btype="bandstop", analog=False, output="ba", fs=10_000))

subtitle = ["lowpass", "highpass", "bandpass", "bandstop"]
fig = plt.figure(figsize=(10, 8), dpi=130)
outer = gridspec.GridSpec(2, 2)

for i in range(4):
    inner = gridspec.GridSpecFromSubplotSpec(2, 1, subplot_spec=outer[i])

    b, a = ba[i]
    filtered = signal.filtfilt(b, a, sig)

    ax = plt.Subplot(fig, inner[0])
    fig.autofmt_xdate(rotation=45)
    ax.set_title(subtitle[i])
    ax.plot(t, filtered)
    ax.plot(t, sig, alpha=0.5)
    fig.add_subplot(ax)

    ax = plt.Subplot(fig, inner[1])
    fig.autofmt_xdate(rotation=45)
    x_ = fft.fftfreq(len(filtered), 1/300)
    y_ = np.abs(fft.fft(filtered)) / len(t)
    x, y = zip(*sorted(zip(x_, y_)))
    ax.plot(x, y)
    fig.add_subplot(ax)
fig.tight_layout()
````
